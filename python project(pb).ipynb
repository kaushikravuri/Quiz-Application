```python
import tkinter                                                        #Importing Tkinter module........
from tkinter import *                                                 #Importing everything from Tkinter module........
import random                                                         #Importing Random module.........


questions = [                                                         #This list used to store the Questions.......
    "r = lambda q: q * 2\ns = lambda q: q * 3\nx = 2\nx = r(x) \nx = s(x) \nx = r(x) \nprint x ",
    "a = True\nb = False\nc = False\n\nif a or b and c: \n    print ""Dr.SUKHVIR KAUR MAM""\nelse: \n    print ""Dr.Sukhvir Kaur mam""",
    "count = 1 \n\ndef doThis(): \n\n    global count \n\n    for i in (1, 2, 3):  \n        count += 1\n\ndoThis() \n\nprint count ",
    "Which of The Following is must to Execute a Python Code ?",
    "from random import randrange \nL = list() \nfor x in range(5): \n    L.append(randrange(0, 100, 2)-10) \n\n# Choose which of outputs below are valid for this code.     \n\nprint(L) ",
    "The output of executing string.ascii_letters can also be achieved by:",
    "If a=(1,2,3,4), a[1:-1] is _________",
    "Which of the following statements create a dictionary?",
    "Which module in the python standard library parses options received from the command line?",
    "To Declare a Global variable in python we use the keyword ?",
]

answers_choice = [                                                   #This list used to store options.......
    ["A)   23","B)  24","C)  25","D)  32",],
    ["A)  Dr.Sukhvir Kaur mam","B)  Dr.SUKHVIR KAUR MAM","C)  Syntex Error","D)  Both A and B",],
    ["A)  2","B)  3","C)  4","D)  5",],
    ["A)  TURBO C","B)  Py Interpreter","C)  Notepad","D)  IDE",],
    ["A)  [-8, 88, 8, 58, 0]","B)  [-8, 81, 18, 46, 0]","C)  [-7, 88, 8, 58, 0]","D)  [-8, 88, 94, 58, 0]",],
    ["A)  string.ascii_lowercase_string.digits","B)  string.ascii_lowercase+string.ascii_upercase","C)  string.letters","D)  string.lowercase_string.upercase",],
    ["A)  Error, tuple slicing doesn’t exist","B)  [2,3]","C)  (2,3,4)","D)  (2,3)",],
    ["A)  d = {}","B)  d = {“john”:40, “peter”:45}","C)  d = {40:”john”, 45:”peter”}","D)  All of the mentioned",],
    ["A)  main","B)  os","C)  getarg","D)  getopt",],
    ["A)  all","B)  var","C)  let","D)  global",],
] 

answers = [1,1,2,1,3,1,3,3,0,3]                                             #list of Correct answers......

user_answer = []                                                   #Declaring a variable to store user answers......

indexes = []                                            #Declare the array variable to select the index in array.......


def gen():                                          #Declare this function to select random indexes in array(MCQ's)........
    global indexes
    while(len(indexes) < 10):                        #This loop will allows only 10 questions to display on screen........
        x = random.randint(0,9)                 #imported random for this it will select random in a given array........
        if x in indexes:
            continue
        else:
            indexes.append(x)


def showresult(score):                                  #this function is declared to show score on screen........
    lblQuestion.destroy()
    r1.destroy()
    r2.destroy()                                    #In this whole code distroy will helps you to distroy a widget.......
    r3.destroy()
    r4.destroy()
   
    labelresulttext = Label(                           #Label uses to design a text and background........
        root,
        font = ("Consolas",20),
        background = "#1A1A1D",
        fg="#FACA2F"
    )
    labelresulttext.pack(pady=(200,30))
    
    labelresult1text = Label(
        root,
        font = ("Consolas",20),
        background = "#1A1A1D",
        fg="#FACA2F"
    )
    labelresult1text.pack(pady=(30,0))
    
    labelresult2text = Label(
        root,
        font = ("Consolas",20),
        background = "#1A1A1D",
        fg="#FACA2F"
    )
    labelresult2text.pack()                                       #It will display the score on the screen........
    if score == 30:
        labelresulttext.configure(text="YOU ARE OUT STANDING")
        
    elif score >= 24:
        labelresulttext.configure(text="You Are Excellent !!")
        
    elif (score >= 13 and score < 24):
        labelresulttext.configure(text="You Can Be Better !!")
        
    else:
        labelresulttext.configure(text="You Should Work Hard !!")
        
    labelresult1text.configure(text="Your Score out of 30:")
    labelresult2text.configure(text=score,)
        
    print("score =",score)


def calc():                                        #This function is declared to calculate the score of the user........
    global indexes,user_answer,answers
    x = 0
    score = 0
    for i in indexes:                             #This loop will send the indexes of Questions into i........
        if user_answer[x] == answers[i]:          #It will tally the user answers and correct answers........
            score = score + 3
        x += 1
    print(score)
    showresult(score)


ques = 1
def selected():                           #This function is declared to collects the user input in radiobuttons.........
    global radiovar,user_answer
    global lblQuestion,r1,r2,r3,r4
    global ques
    x = radiovar.get()
    user_answer.append(x)
    
    radiovar.set(-1)                              #NO USE OF THIS...........
    
    if ques < 10:
        lblQuestion.config(text= questions[indexes[ques]])
        r1['text'] = answers_choice[indexes[ques]][0]
        r2['text'] = answers_choice[indexes[ques]][1]
        r3['text'] = answers_choice[indexes[ques]][2]
        r4['text'] = answers_choice[indexes[ques]][3]
        ques += 1
    else:
        print(indexes)
        print(user_answer)
        calc()
    




def startquiz():                                 #This function is declared to display the MCQ's.........
    global lblQuestion,r1,r2,r3,r4
    lblQuestion = Label(
        root,
        text = questions[indexes[0]],
        font = ("Consolas", 16),
        width = 500,
        justify = "left",
        wraplength = 400,
        background = "#1A1A1D",
        fg="#C3073F",
    )
    lblQuestion.pack(pady=(50,30))

    global radiovar
    radiovar = IntVar()                       #Declaring variable to store radiobutton inputs.........
    radiovar.set(-1)

    r1 = Radiobutton(            #In this whole code radiobutton is declared to display the options in radiobutton view....
        root,
        text = answers_choice[indexes[0]][0],
        font = ("Times", 12),
        value = 0,
        variable = radiovar,
        command = selected,
        background = "#1A1A1D",
        fg="white",
    )
    r1.pack(pady=5)

    r2 = Radiobutton(
        root,
        text = answers_choice[indexes[0]][1],              #Here 0 and 1 indicates the option in answer_choice........
        font = ("Times", 12),
        value = 1,
        variable = radiovar,
        command = selected,
        background = "#1A1A1D",
        fg="white",
    )
    r2.pack(pady=5)

    r3 = Radiobutton(
        root,
        text = answers_choice[indexes[0]][2],
        font = ("Times", 12),
        value = 2,
        variable = radiovar,
        command = selected,
        background = "#1A1A1D",
        fg="white",
        justify = "left",
    )
    r3.pack(pady=5)

    r4 = Radiobutton(
        root,
        text = answers_choice[indexes[0]][3],
        font = ("Times", 12),
        value = 3,
        variable = radiovar,
        command = selected,
        background = "#1A1A1D",
        fg="white",
        justify = "left",
    )
    r4.pack(pady=5)


def startIspressed():                        #This function is declared to start quiz........
    labeltext.destroy()
    lblInstruction.destroy()
    lblInstruction1.destroy()
    lblRules.destroy()
    btnStart.destroy()
    gen()
    startquiz()
 


root = tkinter.Tk()                             #Declaring Tkinter in this code.........
root.title("Quizrocks")
root.geometry("700x600")                        #This use for the dimensions for tkinter........
root.config(background="#1A1A1D")
root.resizable(0,0)

labeltext = Label(
    root,
    text = "QUIZ",
    font = ("Comic sans MS",40,"bold"),
    background = "#1A1A1D",
    fg="#C3073F"
)
labeltext.pack(pady=(0,50))                     #This used to change the widgets........


btnStart = Button(                              #Button is declared to start quiz........
    root,
    text="START",
    relief = FLAT,
    font=("Fraktur Font",15,"bold"),
    border = 0,
    bg="#6F2232",
    fg="white",
    command = startIspressed,
)
btnStart.pack(pady=80)

lblInstruction = Label(                             #This label is declared to display instructions........
    root,
    text = "Read The Rule and know about The Quiz.\n",
    background = "#1A1A1D",
    fg="#C3073F",
    font = ("Consolas",14),
    justify = "center",
)
lblInstruction.pack(pady=(10,30))

lblRules = Label(                                 #This label is declared to display rules on screen........
    root,
    text = "There are 10 Questions.\nMust and should attempt every Question.\nNo negative marks\nOnly one answer for one question\nJust click "
    "one option to attempt the question.",
    width = 100,
    font = ("Times",14),
    background = "#4E4E50",
    foreground = "#FACA2F",
    
)
lblRules.pack()

lblInstruction1 = Label(                             #This label is declared to display instructions........
    root,
    text = "Click START Once You Are ready.",
    background = "#1A1A1D",
    fg="#C3073F",
    font = ("Consolas",20),
    justify = "center",
)
lblInstruction1.pack(pady=(10,0))

root.mainloop()                                     #This is declared to run the tkinter(MUST)........
```
